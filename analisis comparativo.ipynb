Pregunta 2

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url_tesla = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_data = requests.get(url_tesla).text
soup = BeautifulSoup(html_data, "html.parser")

tables = soup.find_all("table")

# Imprimir todas las tablas encontradas para inspeccionarlas
print(f"Se encontraron {len(tables)} tablas en la página.")
for idx, table in enumerate(tables):
    print(f"\nTabla {idx+1}:")
    print(table.prettify())  # Mostrar el HTML bien estructurado

# Buscar la tabla de ingresos trimestrales de Tesla
tesla_revenue = None
for table in tables:
    if "Tesla Quarterly Revenue" in str(table):
        tesla_revenue = pd.read_html(str(table))[0]
        break

# Comprobar si tesla_revenue fue definida
if tesla_revenue is not None:
    print(tesla_revenue.tail())
else:
    print("No se encontró la tabla de ingresos trimestrales de Tesla.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url_gme = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
html_data = requests.get(url_gme).text
soup = BeautifulSoup(html_data, "html.parser")

tables = soup.find_all("table")

for table in tables:
    if "GameStop Quarterly Revenue" in str(table):
        gme_revenue = pd.read_html(str(table))[0]
        break

# Mostrar las últimas cinco filas
print(gme_revenue.tail())
```
